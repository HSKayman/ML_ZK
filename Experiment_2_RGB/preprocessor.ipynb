```python
import os
from PIL import Image
```

```python
input_folders = ["raw_data/cat/", "raw_data/dog/", "raw_data/squirrel/"]
output_folders = ["preprocessed_data_RGB/cat/", "preprocessed_data_RGB/dog/","preprocessed_data_RGB/squirrel/"]

for i in output_folders:
    os.makedirs(i, exist_ok=True)
```

```python
for index, folders in enumerate(input_folders):
    counter = 1
    print(f"{output_folders[index].split('/')[1]} : {len(os.listdir(folders))} images")
```

```python
for index, folders in enumerate(input_folders):
    counter = 1
    for i, filename in enumerate(os.listdir(folders)):
        if filename.endswith((".jpg", ".jpeg", ".png")) and i < 1600:  # Adjust extensions as needed
            # Load the image
            img_path = os.path.join(folders, filename)
            img = Image.open(img_path)

            # Convert to grayscale
            # img = img.convert("L")

            # Resize to 224x224
            img = img.resize((224, 224))
            
            # Save the image
            new_filename = f"{output_folders[index].split('/')[1]}_{counter}.png"
            counter +=1
            output_path = os.path.join(output_folders[index], new_filename)
            img.save(output_path)
            print(f"Saved {output_path}")
```
